## Завдання 1


```python
import numpy as np
import matplotlib.pyplot as plt

h = np.pi / 20
x = np.arange(-2 * np.pi, 2 * np.pi + h, h)

y = np.cos(x) ** 3
z = np.sin(x) + np.sin(2 * x)

plt.figure(figsize=(10, 6))

plt.plot(x, y, "y*", label=r"$y=\cos^3(x)$")
plt.plot(x, z, "rs", label=r"$z=\sin(x)+\sin(2x)$")

plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.title("Графіки функцій $y=cos^3(x)$ та $z=sin(x)+sin(2x)$", fontsize=14)

plt.grid(True)
plt.legend()

plt.text(-5, 1.5, "Порівняння двох тригонометричних функцій", fontsize=10, color="blue")

plt.show()
```

## Завдання 2


```python
import numpy as np
import matplotlib.pyplot as plt

h = np.pi / 20
x = np.arange(-2 * np.pi, 2 * np.pi + h, h)

y = np.cos(x) ** 3
z = np.sin(x) + np.sin(2 * x)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.plot(x, y, "y*", label=r"$y=\cos^3(x)$")
ax1.set_xlabel("x")
ax1.set_ylabel("y(x)")
ax1.set_title("Графік функції y(x) = cos^3(x)")
ax1.grid(True)
ax1.legend()
ax1.text(-5, 0.8, "Жовта лінія зі зірочками", fontsize=9, color="blue")

ax2.plot(x, z, "rs", label=r"$z=\sin(x)+\sin(2x)$")
ax2.set_xlabel("x")
ax2.set_ylabel("z(x)")
ax2.set_title("Графік функції z(x) = sin(x)+sin(2x)")
ax2.grid(True)
ax2.legend()
ax2.text(-5, 1.5, "Червона лінія з квадратами", fontsize=9, color="blue")

fig.suptitle("Графіки функцій y(x) та z(x)", fontsize=14)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

## Завдання 3


```python
import numpy as np
import matplotlib.pyplot as plt

h = np.pi / 100
theta = np.arange(0, 2 * np.pi + h, h)
r = np.sin(3 / 2 * theta)

plt.figure(figsize=(6, 6))
ax = plt.subplot(polar=True)
ax.plot(theta, r, "m-", linewidth=2, label="r = sin(3/2·x)")
ax.set_title("Полярний графік функції r = sin(3/2·x)", va="bottom")
ax.legend(loc="upper right")
plt.show()
```

## Завдання 4


### Частина 1

```python
import numpy as np
import matplotlib.pyplot as plt

h = np.pi / 50
x = np.arange(-2 * np.pi, 2 * np.pi + h, h)
y = np.arange(-2 * np.pi, 2 * np.pi + h, h)

X, Y = np.meshgrid(x, y)

Z = np.exp(X) + np.exp(Y) - 2 ** (X * Y) - 1

plt.figure(figsize=(7, 6))
contour = plt.contour(X, Y, Z, levels=[0], colors="red")
plt.xlabel("x")
plt.ylabel("y")
plt.title(r"Неявний графік: $e^x + e^y - 2^{xy} - 1 = 0$")
plt.grid(True)
plt.show()
```

### Частина 2

```python
import numpy as np
import matplotlib.pyplot as plt

h = np.pi / 20
t = np.arange(-2 * np.pi, 2 * np.pi + h, h)

x = 2 * np.cos(t) ** 2 + 8 * np.cos(t)
y = np.sin(2 * t) + 8 * np.sin(t)

plt.figure(figsize=(8, 6))
plt.plot(x, y, "b-", linewidth=2, label=r"$x=2\cos^2 t+8\cos t,\ y=\sin 2t+8\sin t$")
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.title("Параметричний графік на інтервалі [-2π, 2π]")
plt.legend()
plt.grid(True)
plt.axis("equal")
plt.show()
```

### Частина 3

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)
Z = np.exp(X) + np.exp(Y) - 2**(X*Y) - 1

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("Спосіб 3: 3D поверхня (аналог ezmesh)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.show()
```

### Частина 4 (contour)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.2)
y = np.arange(-5, 5, 0.2)
X, Y = np.meshgrid(x, y)
Z = np.exp(X) + np.exp(Y) - 2**(X*Y) - 1

fig, ax = plt.subplots(figsize=(7,6))

contour = ax.contour(X, Y, Z, levels=30, cmap='jet')
ax.clabel(contour, inline=True, fontsize=8)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Аналог ezcontour")

plt.grid(True)
plt.show()
```
