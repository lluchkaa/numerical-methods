## Завдання 1

Вхідні дані

```python
import sympy as sp

x = sp.Array([-7, -4, -1, 2, 5])
y = sp.Array([-8.5, 3.5, -2.5, 4.5, -3.5])
```

### Звичайний многочлен

Матриця А:

```python
import sympy as sp

A = sp.Matrix([[val**i for i in range(len(x))] for val in x])

A
```

Матриця Б:

```python
import sympy as sp

B = sp.Matrix([[v] for v in y])

B
```

Розвʼязок системи рівнянь:

```python
import sympy as sp

X = A**-1 * B

X
```

```python
import matplotlib.pyplot as plt
import numpy as np

h = 0.1

x1 = np.arange(-10, 10 + h, h)
y1 = sum(X[i] * x1**i for i in range(len(X)))

plt.figure(figsize=(8, 6))
plt.plot(x, y, "*b", markersize=10, label="Висота")
plt.plot(x1, y1, "r", label="Звичайний многочлен")
plt.title("Графік ")
plt.xlabel("Вік дерев, років")
plt.ylabel("Висота, м")
plt.grid(True)
plt.legend()
plt.show()
```

## Завдання 2

Вхідні дані

```python
import numpy as np

X = np.array([10, 30, 50, 70, 90, 110, 130])
Y = np.array([4.2, 15.2, 19.2, 22.4, 24.1, 27.1, 28.3])
```

Точковий графік

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(X, Y, "*b", markersize=10, label="Висота")
plt.title("Графік розміщення точок (вік - висота дерев)")
plt.xlabel("Вік дерев, років")
plt.ylabel("Висота, м")
plt.grid(True)
plt.legend()
plt.show()
```

Графік апроксимації

```python
import numpy as np
import matplotlib.pyplot as plt

X1 = np.log(X)

S1 = np.sum(X1)
S2 = np.sum(X1**2)

A = np.array([[S2, S1], [S1, len(X)]])

b1 = np.sum(X1 * Y)
b2 = np.sum(Y)
B = np.array([b1, b2])

K = np.linalg.solve(A, B)
k1, k0 = K  # коефіцієнти апроксиманти

print(f"Формула апроксиманти: y = {k1:.4f} * ln(x) + {k0:.4f}")

x_vals = np.linspace(10, 130, 200)
y_vals = k1 * np.log(x_vals) + k0

plt.scatter(X, Y, color="red", label="Вихідні дані")
plt.plot(x_vals, y_vals, color="blue", label="Апроксиманта")
plt.grid(True)
plt.title("Графік апроксиманти")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
